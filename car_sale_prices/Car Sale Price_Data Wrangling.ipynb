## Deal with missing data

How to deal with missing data?
drop data
a. drop the whole row
b. drop the whole column
replace data
a. replace it by mean
b. replace it by frequency
c. replace it based on other functions
Whole columns should be dropped only if most entries in the column are empty. In our dataset, none of the columns are empty enough to drop entirely. We have some freedom in choosing which method to replace data; however, some methods may seem more reasonable than others. We will apply each method to many different columns:

Replace by mean:

"normalized-losses": 41 missing data, replace them with mean
"stroke": 4 missing data, replace them with mean
"bore": 4 missing data, replace them with mean
"horsepower": 2 missing data, replace them with mean
"peak-rpm": 2 missing data, replace them with mean
Replace by frequency:

"num-of-doors": 2 missing data, replace them with "four".
Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur
Drop the whole row:

"price": 4 missing data, simply delete the whole row
Reason: price is what we want to predict. Any data entry without price data cannot be used for prediction; therefore any row now without price data is not useful to us

### Replacing Missing Values with mean

```python
# calculating average of column 'normalized-losses'
avg_nor_loss = df['normalized-losses'].astype(float).mean(axis=0)
print(avg_normalized_losses)
#replacing missing values with mean value
df['normalized-losses'].replace(np.nan, avg_nor_loss, inplace=True)
```

```python
# calculating average of column 'stroke'
stroke = df['stroke'].astype(float).mean(axis=0)
print(stroke)
#replacing missing values with mean value
df['stroke'].replace(np.nan, stroke, inplace=True)
```

```python
# calculating average of column 'bore'
bore = df['bore'].astype(float).mean(axis=0)
print(bore)
#replacing missing values with mean value
df['bore'].replace(np.nan, bore, inplace=True)
```

```python
# calculating average of column 'horsepower'
hor_pow = df['horsepower'].astype(float).mean(axis=0)
print(hor_pow)
#replacing missing values with mean value
df['horsepower'].replace(np.nan, hor_pow, inplace=True)
```

```python
# calculating average of column "peak-rpm"
peak = df["peak-rpm"].astype(float).mean(axis=0)
print(peak)
#replacing missing values with mean value
df["peak-rpm"].replace(np.nan, peak, inplace=True)
```

### Replacing Missing Values with frequency

```python
# counts values in column 'num-of-doors'
df['num-of-doors'].value_counts()
# identify max value in column 'num-of-doors'
df['num-of-doors'].value_counts().idxmax()
#replacing missing values with mean value
df['num-of-doors'].replace(np.nan, "four", inplace=True)
```

### Data Normalization

Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include scaling the variable so the variable average is 0, scaling the variable so the variance is 1, or scaling variable so the variable values range from 0 to 1


```python
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

### Binning
Why binning?
Binning is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis.

Example:

In our dataset, "horsepower" is a real valued variable ranging from 48 to 288, it has 57 unique values. What if we only care about the price difference between cars with high horsepower, medium horsepower, and little horsepower (3 types)? Can we rearrange them into three â€˜bins' to simplify analysis?


```python
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(df['horsepower'])


# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
plt.show()
```

```python
#creating the bins
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
#labeling the bins
group_names = ['Low', 'Medium', 'High']
```

```python
#creating a new column with bin names
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names,  include_lowest=True )
df[['horsepower','horsepower-binned']].head(10)
```

```python
#plotting with new bins with bar chart
%matplotlib inline
import matplotlib.pyplot as plt
plt.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
plt.show()
```

```python
# plotting traditional histogram 
%matplotlib inline
import matplotlib.pyplot as plt

a = (0,1,2)

# draw historgram of attribute "horsepower" with bins = 3
plt.hist(df["horsepower"], bins = 3)

# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
plt.show()
```

### indicator variable (or dummy variable)
An indicator variable (or dummy variable) is a numerical variable used to label categories. They are called 'dummies' because the numbers themselves don't have inherent meaning. 

```python
df.columns
```

```python
# creating dummy values for fuel column
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
df.head()
```

```python
# creating dummy values for aspiration column
dummy_variable_2 = pd.get_dummies(df["aspiration"])
# change column names for clarity
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)
```

```python
# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_variable_2], axis=1)

# drop original column "fuel-type" from "df"
df.drop("aspiration", axis = 1, inplace=True)
df.head()
```
