# Gas Price Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('gas_prices.csv')
```

```python
df
```

```python
plt.figure(figsize=(10,8))
plt.plot(df.Year, df['France'], 'r--.')
plt.plot(df.Year, df['Italy'], 'g', marker='.')

plt.xticks(df.Year[::2], rotation=45)
plt.grid()
plt.xlabel('Year', size=12, weight='bold')
plt.ylabel('US Dollars ($)', size=12, weight='bold')
plt.title('Gas Price Over Time - France Vs Italy', fontdict={'fontweight':'bold', 'fontsize':18}, color='r')
labels= ['France', 'Italy']
plt.legend(labels)
plt.show()
```

# Plot all countries

```python


plt.figure(figsize=(10,8))
for country in df:
    if country != 'Year':
        plt.plot(df['Year'], df[country], marker='.')

plt.xticks(df.Year[::2].to_list() + [2010], rotation=45, ha='right')
plt.xlabel('Year', size=12, weight='bold')
plt.ylabel('US Dollars ($)', size=12, weight='bold')
plt.title('Gas Price Over Time', fontdict={'fontweight':'bold', 'fontsize':18}, color='r')
labels=['Australia', 'Canada', 'France', 'Germany', 'Italy', 'Japan', 'Mexico', 'South Korea', 'UK', 'US']
plt.legend(labels)

plt.show()
```
